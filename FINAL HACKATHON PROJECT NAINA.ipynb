The skewness value of 0.026 indicates a very small(negligible) amount of right skewness. Therefore, the 'Age' variable can be considered approximately symmetrical or normally distributed.

```python
# Pair plot grouped by the only categorial variable
sns.pairplot(train, hue = 'Healthy', height = 2)
```

The pairplot clearly shows the columns "Mental Health Management" and "Illness Count last year" as unbalanced . Let's explore them . It seems most of the features are highly correlated with others.
We can see there are several outliers in the "Illness count last year" , we should remove them

```python
count_ones = train["Mental health management"].value_counts()[1]

# Print the count
print("Number of 1's in 'Mental health management':", count_ones)
```

```python
count_ones = test["Mental health management"].value_counts()[1]

# Print the count
print("Number of 1's in 'Mental health management':", count_ones)
```

The occurence of 1 in both datasets is rare and later in feature scores , we see the score of this column is extremely low , so it's better to remove it to avoid any adverse effects . (In measuring feature importance of the model also , we find that it scores very low)

```python
train.drop('Mental health management',axis = 1 , inplace = True)
test.drop('Mental health management',axis = 1 , inplace = True)
```

```python
train.shape
```

```python
Q1 = train['Illness count last year'].quantile(0.25)
Q3 = train['Illness count last year'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = train[(train['Illness count last year'] < lower_bound) | (train['Illness count last year'] > upper_bound)]

print(outliers)
```

```python
train = train.drop(train.index.intersection(outliers))
print(train.shape)
```

# Encoding some categorical features manually 

```python
train['Smoker?'].unique()
```

```python
train['Smoker'] = train['Smoker?'].apply(lambda x : 1 if x=='YES' else (0 if x=='NO' else  2))
```

```python
train['Living in?'].unique()
```

```python
train['Area'] = train['Living in?'].apply(lambda x:1 if x == 'RURAL' else 0)
```

```python
train.shape
```

```python
train['Follow Diet'].unique()
```

```python
train['Follow Diet'] = train['Follow Diet'].apply(lambda x : 1 if x == 1.0 else 0)
```

# DATA CLEANING

"Any heriditary condition?" is a constant feature .So ,it has no effect on predictions. Therefore , it should be removed .

# Now , remove null values

```python
train.isnull().sum()
```

```python
train['Food preference'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming you have a pandas DataFrame called 'data' with a column 'Food preference'
food_preference = train['Food preference']

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the 'Food preference' column
encoded_food_preference = label_encoder.fit_transform(food_preference)

# Assign the encoded values back to the DataFrame
train['Food preference'] = encoded_food_preference
```

```python
train.isnull().sum()
```

We need to check which null values are common

```python
nav = train[train.isna().any(axis=1)]
print(nav)
```

Clearly , all these rows share missing data across all columns , so we can drop these rows . These rows are only 3.75% of the whole data . So no effect

```python
train.dropna(inplace = True) #Drop null values permanently
print(train.isnull().sum())
print(train.shape)
```

Now , we drop the columns with 'string' values as we already have encoded them

```python
df = train.drop(['Smoker?','Living in?'],axis = 1)
```

```python
df = df.drop(['Taking supplements','ID1','ID2'],axis = 1) #BASED ON FEATURE IMPORTANCE SCORES
```

```python
df.head()
```

```python
df.shape
```

```python
train['Follow Diet'] = np.log(train['Follow Diet'])
train['Physical activity'] = np.log(train['Physical activity'])
train['Alcohol consumption'] = np.log(train['Alcohol consumption'])
```
